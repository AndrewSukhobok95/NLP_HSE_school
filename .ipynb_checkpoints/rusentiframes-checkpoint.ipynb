```python
import pandas as pd
import os.path
import re
```

```python
#
with open('rusentiframes_copy.txt', 'r', encoding = 'utf-8') as f_sentiframes:
    entries = []
    keys = []
    entry = []
    spattern = re.compile(r'^\s+')
    for  line in f_sentiframes.readlines():
        if line == '\n':
            if len(entry) != 0:
                entries.append(entry)
                entry = []
        else:
            if re.fullmatch(spattern,line) != None:
                if len(entry) != 0:
                    entries.append(entry)
                    entry = []
            else:
                entry.append(line.strip('\n'))
                
```

```python
for entry in entries:
    keys.append(entry[0].split()[0].lower())
```

```python
entries_upd = []
keys_new=[]
values_new=[]
for entry in entries:
    entry_upd = {}
    for i in range(len(entry)):
        if i != 0:
            line = entry[i].split(maxsplit=1)
            key = line[0].strip(':')
            key = key.strip(',')
            value = line[1]
            if entry_upd.get(key) == None:
                entry_upd[key] = value
            else:
                if isinstance(entry_upd[key], str) == True:
                    value = [entry_upd[key],value]
                    entry_upd[key] = value
                elif isinstance(entry_upd[key], list) == True:
                    value = entry_upd[key].append(value)
    entries_upd.append(entry_upd)
    
```

```python
sentiframes = dict(zip(keys, entries_upd)) 
```

```python
sentiframes_df = pd.DataFrame.from_dict(sentiframes, orient = 'index')
```

```python
#rearranging columns
cols = sentiframes_df.columns.tolist()
cols = cols[2:3] + cols[4:] + cols[0:2] + cols[3:4]
```

```python
sentiframes_df = sentiframes_df[cols]
```
